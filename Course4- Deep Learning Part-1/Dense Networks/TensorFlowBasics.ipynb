```python
import tensorflow as tf
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

## Types of tensor

- **tf.Constant:** Immutable tensor ie. can not be changed after the declaration
- **tf.Variable:** It can be changed after the decalaration

## Decalare a tensor

```python
# constant tensor

tf_const = tf.constant([1,2,3], dtype='int32')
tf_const
```

```python
# to print the element of tensor we will use numpy() function

tf_const.numpy()
```

```python
# variable tensor

tf_var = tf.Variable([1,2,3], dtype='int32')
tf_var
```

```python
tf_var.numpy()
```

**Que)** Declare a tensor of shape (2, 2, 2, 2) with random numbers from a normal distribution whose mean and standard deviation are both 1.

```python
# approach 1: using numpy array

arr = np.random.randn(2, 2, 2, 2) # it generates a random array
arr = arr + 1
tf_arr = tf.Variable(arr, dtype='float64')
tf_arr
```

```python
sns.distplot(tf_arr.numpy().reshape(16,1))
plt.show()
```

```python
# approach 2: using tensor library

tf_arr = tf.random.normal([2, 2, 2, 2], mean=1, stddev=1)
tf_arr
```

```python
sns.distplot(tf_arr.numpy().reshape(-1,1))
plt.show()
```
