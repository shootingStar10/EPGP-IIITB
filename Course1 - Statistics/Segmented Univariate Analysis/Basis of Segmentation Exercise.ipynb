```python
import pandas as pd
```

```python
data = pd.read_csv('EDA_nas.csv')
data.head()
```

```python
# What is the impact of the variable 'Watch.TV' on Science marks?
data = data[~data['Science..'].isnull()]
data.groupby(by = 'Watch.TV')['Science..'].describe()
```

```python
#What is the impact of the father’s education on the student’s Maths marks?
data = pd.read_csv('EDA_nas.csv')
data = data[~data['Maths..'].isnull()]
data.groupby(by = 'Father.edu')['Maths..'].describe()
```

```python
# What is the impact of 'Play.Games' on Reading marks?
data = pd.read_csv('EDA_nas.csv')
data = data[~data['Reading..'].isnull()]
data.groupby(by = 'Play.games')['Reading..'].describe()
```
