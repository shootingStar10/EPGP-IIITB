```python
# import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import dataset of tendulkar's batting
data = pd.read_csv('tendulkar_ODI.csv')
data.head()
```

```python
# set index as first column
data.set_index('S.No.', inplace=True)
data.head()
```

```python
# get data description
data.describe()
```

```python
# get data info
data.info()
```

```python
# clean run column by removing irrelevant string
data = data[~data['Runs'].apply(lambda x: x in ['DNB', 'TDNB'])]
```

```python
data['Runs'] = data['Runs'].apply(lambda x: x.replace('*', ''))
```

```python
# change datatype of runs columns as int
data['Runs'] = data['Runs'].astype('int32')
```

```python
# plot histogram 
sns.histplot(data = data, x = 'Runs', bins=20)
plt.show()
```

```python
# set datatype of column 4s as int
data['4s'] = data['4s'].astype('int32')
```

```python
# plot a bar chart of 4s and its frequency
sns.countplot(data = data, x = '4s')
plt.show()
```
