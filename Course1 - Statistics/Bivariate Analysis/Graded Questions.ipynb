#### Q2) You also want to understand how the average science marks vary with the father’s education level and age. Neglect the age group of 11 years since there are very few children in that segment. Mark the correct statement:

```python
# remove null values in science marks column
X = data[~data['Science..'].isnull()]
```

```python
# remove null values in father's education column
X = X[~X['Father.edu'].isnull()]
```

```python
# remove values of 11- years from age column
X = X[X['Age'] != '11- years']
```

```python
# remove null values in age column
X = X[~X['Age'].isnull()]
```

```python
# filter data based on required columns
filter_data = X.loc[:, ['Science..', 'Father.edu', 'Age']]
```

```python
# plot bar chart wrt to average of science marks and categories as 'Father.edu' and 'Age'
filter_data.groupby(by = ['Father.edu', 'Age'])['Science..'].mean().plot.bar()
plt.show()
```
