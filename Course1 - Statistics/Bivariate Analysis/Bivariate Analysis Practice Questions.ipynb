#### Plot a correlation matrix for Euro, Japanese Yen, U.K. Pound Sterling, US Dollar, Chineese Yuan, Australian Dollar and Indian Rupee.

```python
filter_data = data.loc[:, ['Euro', 'Japanese Yen', 'U.K. Pound Sterling', 'U.S. Dollar', 'Chinese Yuan', 'Australian Dollar', 'Indian Rupee']]
```

```python
filter_data
```

```python
filter_data = filter_data[~filter_data['Indian Rupee'].isnull()]
```

```python
filter_data = filter_data[~filter_data['Japanese Yen'].isnull()]
```

```python
filter_data
```

```python
corr_matrix = filter_data.corr()
```

```python
sns.heatmap(data = corr_matrix, cmap = 'Greens', annot=True)
plt.show()
```
